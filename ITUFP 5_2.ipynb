```python
root_database = list()
with open('data/chess_negative.txt', 'r') as file:
    count = 0
    for line in file:
        temp_struc = line.split(":")
        temp_items = list(map(lambda x: "item " + x, temp_struc[0].split(" ")))
        temp_transaction = {
            "tid": count,
            "items": temp_items,
            "quantities": [1] * len(temp_items),
            "profits": list(map(int, temp_struc[2].split(" ")))
        }
        root_database.append(temp_transaction)
        count += 1
```

```python
database = []
for i in root_database:
    database.append({
        'tid': i['tid'],
        'items': i['items'], 
        'profits': [qua*prof for qua,prof in zip(i['quantities'],i['profits'])], 
    })
database
```

#Define UPlist and IMCUPList

```python
class UPList:
  def __init__(self, item):
    self.item = {item}
    self.table = {"TID": [], "Prob": []}
    self.utility = 0
    self.max = -sys.float_info.max
  
  def is_maxper_and_avg_per(self, numTransaction, maxper, maxavg):
    ps = []
    # print("item", self.item)
    # print("HERE", self.table)
    if len(self.table["TID"]) != 0:
      ps.append(self.table["TID"][0])
      for i in range(1,len(self.table["TID"])):
        
        ps.append(self.table["TID"][i]-self.table["TID"][i-1])
        # print("check", self.table["TID"][i], ps)
    if self.table["TID"][-1] != numTransaction:
      ps.append(numTransaction-self.table["TID"][-1])
    
    item_avg_per = sum(ps) / len(ps)
    # print("ps", ps, item_avg_per)
    if max(ps) > maxper or item_avg_per > maxavg:
      return False
    return True
  
  def is_periodic(self, numTransaction, maxper, minper, minavg , maxavg):
    ps = []
    if len(self.table["TID"]) != 0:
      ps.append(self.table["TID"][0])
      for i in range(1,len(self.table["TID"])):
        ps.append(self.table["TID"][i]-self.table["TID"][i-1])
      if self.table["TID"][-1] != numTransaction:
        ps.append(numTransaction-self.table["TID"][-1])
      item_avg_per = sum(ps) / len(ps)
      if max(ps) > maxper or min(ps) < minper or item_avg_per > maxavg or item_avg_per < minavg:
        return False
      return True
    return False

  def __eq__(self, other):
    return self.utility == other.utility
  def __gt__(self, other):
    return self.utility > other.utility
  def __lt__(self, other):
    return self.utility < other.utility

class IMCUPList(UPList):
  def __init__(self, U1, U2):
    UPList.__init__(self, "")
    self.item = U1.item.union(U2.item)
    self.index1 = -1
    self.index2 = -1
    self.U1 = U1
    self.U2 = U2
  def genTableIMCUP(self, minU):
    if self.index2 >= (len(self.U2.table["TID"]) - 1):
      return
    self.index2 += 1
    for self.index1 in range (self.index1 + 1, len(self.U1.table["TID"])):

      while self.index2 < (len(self.U2.table["TID"]) - 1) and self.U1.table["TID"][self.index1] > self.U2.table["TID"][self.index2]:
        self.index2 += 1
        
      if self.U1.table["TID"][self.index1] == self.U2.table["TID"][self.index2]:
        self.table["TID"].append(self.U1.table["TID"][self.index1])
        prob = self.U1.table["Prob"][self.index1] + self.U2.table["Prob"][self.index2]
        self.table["Prob"].append(prob)
        self.utility += prob
        self.max = max(self.max, prob)
        if minU - self.utility > min((len(self.U1.table["TID"]) - self.index1 - 1),(len(self.U2.table["TID"]) - self.index2 - 1)) * (self.U1.max + self.U2.max):
          return
      if self.index2 == len(self.U2.table["TID"]) - 1:
        self.index1 = len(self.U1.table["TID"]) - 1
        return
```

```python
class ITUFPNew:
    def __init__(self, UDB, kList:list, maxper, minper, maxavg, minavg):
        self.kList = kList
        self.maxper = maxper
        self.minper = minper
        self.maxavg = maxavg
        self.minavg = minavg
        self.UDB = UDB
        self.topK = []
        self.IMC =  dict()
        self.pref =  dict()
        self.UPLists = self.genUPList()
    def genUPList(self):
        UPLists = {}
        for transaction in (self.UDB):
            for item, profit in zip(transaction['items'], transaction['profits']):
                if item not in UPLists:
                    UPLists[item] = UPList(item)
                UPLists[item].table['TID'].append(transaction["tid"])
                UPLists[item].table['Prob'].append(profit)  
                UPLists[item].utility += profit
                UPLists[item].max = max(UPLists[item].max, profit)
        finalUPLists = []
        for u in UPLists.values():
            if u.is_maxper_and_avg_per(self.UDB[-1]['tid'], self.maxper, self.maxavg) == True:
                finalUPLists.append(u)
        return sorted(finalUPLists, reverse=True)
    def solve(self):
        for ki in self.kList:
            tk = {"TID": [], "utility": []}
        for i in range(min(ki,len(self.UPLists))):
            tk["TID"].append(self.UPLists[i].item)
            tk["utility"].append(self.UPLists[i].utility)
        minU = -sys.float_info.max
        if len(tk["TID"]) == ki:
            minU =  tk["utility"][-1]

        self.ITUFP_growth(self.UPLists, ki, tk, minU)
        self.topK.append(tk)
    def ITUFP_growth(self, UPLists, ki, tk, minU):
        for indexX, x in enumerate(UPLists[:-1]):
            if x.utility > minU:
              if "".join(x.item) not in self.pref:
                 
                #print("".join(x.item), "not in pref create IMC and pref for", "".join(x.item))
                self.IMC["".join(x.item)] = []
                self.pref["".join(x.item)] = []
              for indexL, l in enumerate(UPLists[indexX+1:]):
                if l.utility > minU:
                  #print("Considered Ul:", "".join(UIMC[l].item))
                  #if x.utility * UIMC[l].max > minSup:
                  if True:
                    index = -1
                    if "".join(x.item.union(l.item)) in self.pref["".join(x.item)]:
                      index = self.pref["".join(x.item)].index("".join(x.item.union(l.item)))
                      if self.IMC["".join(x.item)][index].index1 != len(x.table["TID"]) - 1 and self.IMC["".join(x.item)][index].index2 != len(l.table["TID"]) - 1:
                        #print("Update IMC:",IMC["".join(UIMC[x].item)][index].item)
                        self.IMC["".join(x.item)][index].genTableIMCUP(minU)
                    else:
                      C = IMCUPList(x, l)
                      C.genTableIMCUP(minU)
                      self.IMC["".join(x.item)].append(C)
                      self.pref["".join(x.item)].append("".join(C.item))
                      #print("Create IMC:",IMC["".join(UIMC[x].item)][index].item)
                    if self.IMC["".join(x.item)][index].utility > minU and self.IMC["".join(x.item)][index].is_periodic(self.UDB[-1]["tid"], self.maxper, self.minper, self.minavg , self.maxavg):
                      #print("IMC:\n",IMC)
                      #print("pref:\n",pref)
                      inserted = False
                      for i in range (len(tk["TID"])):
                        if tk["utility"][i] < self.IMC["".join(x.item)][index].utility:
                          tk["TID"].insert(i, self.IMC["".join(x.item)][index].item)
                          tk["utility"].insert(i, self.IMC["".join(x.item)][index].utility)
                          inserted = True
                          break
                      if not inserted:
                        tk["TID"].append(self.IMC["".join(x.item)][index].item)
                        tk["utility"].append(self.IMC["".join(x.item)][index].utility)
                      if len(tk["TID"]) > ki:
                        tk["TID"] = tk["TID"][0:ki]
                        tk["utility"] = tk["utility"][0:ki]
                      if len(tk["TID"]) == ki:
                        minU = tk["utility"][-1]
                      print("TK:\n",tk)
                      print("MinSup:\n",minU)
              if len(self.IMC["".join(x.item)]) > 1:
                self.ITUFP_growth(self.IMC["".join(x.item)],ki,tk,minU)
```
